```python
import pandas as pd

path = './database/user.csv'
avatar_base = 'https://picsum.photos/200/200?random='

# Giả sử DataFrame gốc tên là df
# Lấy user_id và reviewer_name, bỏ trùng
df_user = df[['user_id', 'reviewer_name']].drop_duplicates().copy()

# Đổi tên cột
df_user.rename(columns={'reviewer_name': 'user_name'}, inplace=True)
df_user['full_name'] = df_user['user_name']

# Tạo avatar_url ngẫu nhiên theo user_id (để mỗi user cố định 1 avatar)
df_user['avatar_url'] = df_user['user_id'].apply(
    lambda x: f"{avatar_base}{x}"
)

# Lưu ra CSV
df_user.to_csv(path, index=False)

print(df_user.head())
```

```python
import pandas as pd

df = pd.read_csv('./dulieu_v2.csv')
print('kich thuoc dataset', df.shape)
df.head(3)
```

```python
import ast

def split(data):
    data = ast.literal_eval(data)

    return [data['summary'], data['content']]

df[['summary', 'content']] = df['Thông tin thêm'].apply(
    lambda x: pd.Series(split(x))
)
df.head(3)
```

```python
# place_id,place_name,address,thumb_ulr,description,content
thumburl = 'https://picsum.photos/800/600?random=2'
def get_infor(name):
    df_temp = df[df['Tên địa điểm'] == name]
    if len(df_temp) == 0:
        return ['', thumburl, '', '']
    result = [
        df_temp['Địa chỉ'].values[0],
        thumburl,
        df_temp['summary'].values[0],
        df_temp['content'].values[0]
    ]
    return result

df_place[['address','thumb_ulr','description','content']] = df_place['ten_dia_danh'].apply(
    lambda x: pd.Series(get_infor(x))
)
df_place.head(3)
```

```python
df_place.to_csv('./database/place.csv', index=False)
```
